```python
import logging

# logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("DispAGP")
logger.setLevel(logging.INFO)

# time stamps
if not logger.hasHandlers():
    logging_handler = logging.StreamHandler()
    logging_handler.setFormatter(logging.Formatter("%(asctime)s %(message)s"))
    logger.addHandler(logging_handler)

logger.info("Test")
```

```python
from disp_agp_solver import SatBasedOptimizer, GurobiOptimizer, CpSatOptimizer
from disp_agp_solver import SearchStrategy, OptimizerParams

solver_sat = SatBasedOptimizer(
    instance,
    logger=logger,
    params=OptimizerParams(
        search_strategy_start=SearchStrategy.BINARY,
        search_strategy_iteration=SearchStrategy.DOWN,
        lazy=False,
    ),
)
solver_sat.solve()
print(solver_sat.get_stats())
print(solver_sat.objective)
```

```python
solver_gurobi = GurobiOptimizer(
    instance,
    logger=logger,
)
solver_gurobi.solve(90)
print(solver_gurobi.get_stats())
print(solver_gurobi.objective)
```

```python
solver_cpsat = CpSatOptimizer(
    instance,
    logger=logger,
)
solver_cpsat.solve(90)
print(solver_cpsat.get_stats())
print(solver_cpsat.objective)
```

```python
witness_positions = [w for w, g in solver_sat._witness_strategy.witnesses]
plotter.plot_solution(solver_sat.solution, witness_positions=witness_positions)
plt.show()
```

```python
solver = DispAgpSolver(instance, logger=logger)
# solver.observer = MyObserbver(instance, "example_iterations", "png")
solver.optimize()


print("Guards", solver.guards, "Distance", solver.objective)
```

```python
from pyvispoly import plot_polygon


def plot_solution(ax, instance, objective, guards, closest_pair, title=""):
    ax.set_aspect("equal")
    plt.title(title + "Objective: " + str(objective))
    plot_polygon(instance.as_cgal_polygon(), color="lightblue", ax=ax)
    for guard in guards:
        if guard not in closest_pair:
            ax.plot(
                [instance.positions[guard][0]],
                [instance.positions[guard][1]],
                "bo",
            )
        else:
            ax.plot(
                [instance.positions[guard][0]],
                [instance.positions[guard][1]],
                "ro",
            )


import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 10))
plot_solution(ax, instance, solver.objective, solver.guards, solver.closest_pair)
plt.show()
```
