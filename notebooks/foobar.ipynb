```python
import logging

# logging.basicConfig(level=logging.INFO)
logger = logging.getLogger("DispAGP")
logger.setLevel(logging.INFO)

# time stamps
if not logger.hasHandlers():
    logging_handler = logging.StreamHandler()
    logging_handler.setFormatter(logging.Formatter("%(asctime)s %(message)s"))
    logger.addHandler(logging_handler)

logger.info("Test")
```

```python
from disp_agp_solver import SatBasedOptimizer, GurobiOptimizer, CpSatOptimizer
from disp_agp_solver import SearchStrategy, OptimizerParams

solver_sat = SatBasedOptimizer(
    instance,
    logger=logger,
    params=OptimizerParams(
        search_strategy_start=SearchStrategy.BINARY,
        search_strategy_iteration=SearchStrategy.DOWN,
        lazy=False,
    ),
)
solver_sat.solve()
print(solver_sat.get_stats())
print(solver_sat.objective)
```

```python
solver_gurobi = GurobiOptimizer(
    instance,
    logger=logger,
)
solver_gurobi.solve(90)
print(solver_gurobi.get_stats())
print(solver_gurobi.objective)
```

```python
solver_cpsat = CpSatOptimizer(
    instance,
    logger=logger,
)
solver_cpsat.solve(90)
print(solver_cpsat.get_stats())
print(solver_cpsat.objective)
```

```python
witness_positions = [w for w, g in solver_sat._witness_strategy.witnesses]
plotter.plot_solution(solver_sat.solution, witness_positions=witness_positions)
plt.show()
```
